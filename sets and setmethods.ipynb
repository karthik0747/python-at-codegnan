#### set methods
 - Adding element to set
 - add(ele): it add element to set if element not exists in set,otherwise nothing
 - remove element from set
 - remove (ele): it remove the element from set if element exists in set otherwise
- key error
  - discar(ele):
  - it removes the element from set if element exist in set otherwise nothing
  - set operation
  - union (anthor_set):
  - it return the all commom elements from all sets
  - intersection(anthor_sets):it return the common elements from the all sets
  - difference(anthor_sets):it return these set elements those are not lies in anthor_sets
  - symmetric_difference(anthor):it return the non-common element from all sets


```python
s = {10, 20, 30}
print(s)
s.add(40)
print(s)

s.add(20)   # Already exists, so nothing happens
print(s)
```

```python
s = {10, 20, 30}
print(s)
s.remove(20)
print(s)

# s.remove(50)   # Raises KeyError
```

```python
s = {10, 20, 30}

s.discard(20)
print(s)

s.discard(50)   # No error
print(s)
```

```python
A = {1, 2, 3}
B = {3, 4, 5}

C = A.union(B)

print(C)
```

```python
A = {1, 2, 3}
B = {2, 3, 4}
C = A.intersection(B)
print(C)
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5}
C = B.difference(A)
print(C)
```

```python
A = {1, 2, 3}
B = {3, 4, 5}
C = A.symmetric_difference(B)
print(C)
```
