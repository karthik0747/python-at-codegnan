#### dictionary methods
     -get value
     -get(key,default_val): it returns the value if key exists in dictonary other wise default_val
     -update values and add items and add items to dictionary
      update(new_dictionary):dicitonary updates with new_dictionary items
      -remove items from dictionary
      -pop(key):it return value and remove item if key is exixts in dictionary otherwise 'keyerror'
      -popitea():it returns tuple of key value pair and remove the item from last item from dictionary
      -get keys,value ad items from dictionary
      -keys(): it retuurns list of dictonary keys
      -values():it returns list of dictionary values
      -items(): it return list of tuple key-value pairs

```python
d={'name': 'ravi','age':20,'course':'python','kar':[1,2,3],'age':50}
print(d.get('name'))
print(d.get('location'))
print(d.get('age',18))
print(d.get('location',"location key not exists"))
```

```python
details= { 'name':'ravi', 'age':'20', 'course':'python'}
d = {'marks':[85,90,86],'age':50}
details.update(d)
print(details)
```

```python
d={'name': 'ravi', 'age': 50, 'course': 'python', 'marks': [85, 90, 86]}
print(d.pop('age'))
print(d)
print(d.popitem())
print(d)
```

```python
d={'name': 'ravi', 'age': 50, 'course': 'python', 'marks': [85, 90, 86]}
print(d.keys())
```

```python
details= { 'name':'ravi', 'age':'20', 'course':'python','marks':[85,90,86],'age':50}
keys = details.keys()
print(type(keys))
print(keys)
print(details.values())
print(details.items())
```
