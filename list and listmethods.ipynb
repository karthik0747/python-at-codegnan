#### list methods
 - Elelment adding methods
 - append(item):it adds item at the end of the list it can only one element.
 - extand(itarable):it adds all iterable elements at end of the list
 - insert(pos,item): this method add the item at specified position


```python
lst=[1,2,34,]
lst.append(10)
print(lst)
lst.append([10,20,30])
print(lst)
#extand
lst.extend([11,12,33])
print(lst)
#insert
lst.insert(1,10000)
print(lst)
```

-element remove methods
      -remove(val)
      -pop()
      -pop(pos)
      -del
      -clear()

```python
lst=[10,20,30,20,40]
lst.remove(20)
print(lst)
val=lst.pop()
print(val,lst)
val=lst.pop(1)
print(val,lst)
del lst[0]
print(lst)
lst.clear()
print(lst)
```

```python
# index(val): it returns the index of first occurance of value if value exists in list othewise valueerror
#count(val): it returns the total frequerence of value
lst = [11,2,12,4,14,2,2,2]
ind = lst.index(2)
print(ind)
print(lst.count(2))
print(lst.count(100))
#print(lst.index(100))#ValueError: list.index(x): x not in list
```

```python
#sort(reverse =false) : it sorts the all list elements in ascending order,
#if reverse value is true it sorts in decending order
lst=[1,5,89,7,-1,-9]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
lst1=["ka","abbbbbbbb","cvvvv"]
lst1.sort()
print(lst1)
```
