```python
#indexing finding methods 
#index(sub_string) : it returns the first occurance index of string if exists,if string doesnotexists it gives value error
#find(sub_string)  :it returns the first occurance index of substring if exists,if string doesnotexists it gives value "-1"
string="The ocean is blue and sky is blue"
print(string.index("o"))
print(string.index("is"))
print(string.index("is",14))
#print(string.index("z"))
```

```python
#indexing finding methods 
#index(sub_string) : it returns the first occurance index of string if exists,if string doesnotexists it gives value error
#find(sub_string)  :it returns the first occurance index of substring if exists,if string doesnotexists it gives value "-1"
string="The ocean is blue and sky is blue"
print(string.find("o"))
print(string.find("is"))
print(string.find("is",14))
print(string.find("z"))
print(string.replace('blue','red'))
print(string.replace('blue','red',1))
```

#### string checking methods
- islower(): it returns ture when entire string is in lowercase
- isupper(): it returns ture when entire string is in uppercase
- istitle(): it returns ture when the every word first charcter into upper and remaining to lower
- isalphs(): it returns ture when string has alphabets upper and llower
- isalnum(): it returns true when both alphabets and nums are presents
- isdigit(): it returns true when has only nums
- startswith(substring): it returns true it returns true when string starts with substring
- endswith(substring): it returns true when it ends with substring

```python
print("hello".islower())   
print("HELLO".isupper())    
print("Hello World".istitle())   
print("Hello".isalpha())     
print("Hello123".isalnum())   
print("12345".isdigit())      
print("Python".startswith("Py"))   
print("Python".endswith("on"))  
```
